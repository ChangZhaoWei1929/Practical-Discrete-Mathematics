## Applications to Memory Allocation

### Example: Pre-allocating Memory

Suppose we wish to create a large list of 1,000,000 numbers. The simplest way is to just run a loop, adding one element at a time to the vector. (Note that the runtime will vary depending on the hardware where you run the code.)

````python
import time

number = 1000000

# Check the current time
startTime = time.time()

# Create an empty list
list = []

# Add items to the list one by one
for counter in range(number):
    list.append(counter)

# Display the run time
print(time.time() - startTime)
````

The code below can pre-allocate an array with 1,000,000 and fill it in with 1, 2, ..., 1,000,000. (Note that the runtime will vary depending on the hardware where you run the code.)

````python
import time
number = 1000000

# Check the current time
startTime = time.time()

# Create a list of 1000000 zeros
list = [None]*number

# Add items to the list one by one
for counter in range(number):
    list[counter] = counter

# Display the run time
print(time.time() - startTime)
````

## Efficacy of Brute Force Algorithms

### Example: Caesar Cipher

A brute force check of how the text would be decoded for each possible Caesar cipher.

````python
# Intercepted message
codedMessage = 'nzohfu gur rarzl ng avtug'

# We will shift by 0, shift by 1, shift by 2, ... and print the results
for counter in range(26):
    # Start with no guess
    guessedMessage = ''

    # Loop through each letter in the coded message
    for x in codedMessage:

        # If x is not a space
        if x != ' ':

            # Shift the letter forward by counter
            if ord(x)+counter <= 122:
                x = chr(ord(x)+counter)

            # Subtract 26 if we go beyond z
            else:
                x = chr(ord(x)+counter-26)

        # Build a guess for the message one letter at a time
        guessedMessage = guessedMessage + x

    # Print the counter (the shift) and the message
    print(counter, guessedMessage)
````

Clearly, the length of the shift is 13 letters as only this one produces an intelligible message.

## Generate Random Shift

### Subtask:
Generate a random integer between 1 and 25 (inclusive) to be used as the shift for the Caesar cipher. This will ensure a different encryption each time it runs.


**Reasoning**:
To generate a random integer for the Caesar cipher shift, I need to import the `random` module and then use its `randint()` function.



````python
import random

# Generate a random integer between 1 and 25 (inclusive)
random_shift = random.randint(1, 25)

print(f"Generated random shift: {random_shift}")
````

## Encrypt Message with Random Shift

### Subtask:
Encrypt the message "next week is math exam" using the `caesar_encrypt` function and the randomly generated shift. The `caesar_encrypt` function handles both lowercase and uppercase letters, and preserves non-alphabetic characters.


**Reasoning**:
To encrypt the message, I need to define the `caesar_encrypt` function as specified, which handles character shifting for both uppercase and lowercase letters while preserving non-alphabetic characters. After defining the function, I will apply it to the target message 'next week is math exam' using the `random_shift` value that was generated in the previous step.



````python
def caesar_encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha(): # Check if the character is an alphabet
            start_ascii = ord('A') if char.isupper() else ord('a')
            shifted_char_code = (ord(char) - start_ascii + shift) % 26 + start_ascii
            encrypted_text += chr(shifted_char_code)
        else:
            encrypted_text += char # Preserve non-alphabetic characters
    return encrypted_text

message_to_encrypt = 'next week is math exam'
encrypted_message = caesar_encrypt(message_to_encrypt, random_shift)

print(f"Original message: '{message_to_encrypt}'")
print(f"Chosen random shift: {random_shift}")
print(f"Encrypted message: '{encrypted_message}'")
````

### Decoding an Encrypted Message

To decode the message, we can use a `caesar_decrypt` function, which is essentially the reverse of the encryption process. It shifts each alphabetic character back by the specified number.

````python
def caesar_decrypt(encrypted_text, shift):
    decrypted_text = ''
    for char in encrypted_text:
        if 'a' <= char <= 'z':
            # Handle lowercase letters
            shifted_char_code = ord('a') + (ord(char) - ord('a') - shift) % 26
            decrypted_text += chr(shifted_char_code)
        elif 'A' <= char <= 'Z':
            # Handle uppercase letters
            shifted_char_code = ord('A') + (ord(char) - ord('A') - shift) % 26
            decrypted_text += chr(shifted_char_code)
        else:
            # Keep non-alphabetic characters as they are
            decrypted_text += char
    return decrypted_text

# Using the encrypted message and shift from the previous step
decrypted_message = caesar_decrypt(encrypted_message, random_shift)

print(f"Encrypted message: '{encrypted_message}'")
print(f"Shift amount used for encryption: {random_shift}")
print(f"Decrypted message: '{decrypted_message}'")
````
